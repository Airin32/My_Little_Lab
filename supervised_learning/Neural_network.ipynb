```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_breast_cancer
from sklearn import datasets
import matplotlib.pyplot as plt 
```

```python
x,y = datasets.make_classification(n_features=5,n_classes=2,n_samples=100)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
clf = MLPClassifier(max_iter=5000)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
nln = MLPClassifier(hidden_layer_sizes=(500),activation='logistic',learning_rate='adaptive',learning_rate_init=0.001,warm_start=True,momentum=1.0,max_iter=5000)
```

```python
data_set = load_breast_cancer()
```

```python
df = pd.DataFrame(data=data_set['data'],columns=data_set['feature_names'])
df
```

```python
x = df.copy()
y = data_set['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=4)
```

```python
nln.fit(x_train,y_train)
```

```python
y_pred = nln.predict(x_test)
```

```python
y_pred
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
