```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import Lasso
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data_set = fetch_california_housing(as_frame=True)
```

```python
data_set
```

```python
x = data_set.data
y = data_set.target
```

```python
data_set.target_names
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=4)
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
lass = Lasso()
```

```python
lass.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
y_pred = lass.predict(x_test)
```

```python
y_pred
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
param_grid = {
    'alpha' : [0.0001,0.001,0.1,1,10,100]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lasso_cv = GridSearchCV(lass,param_grid,cv=3,n_jobs=-1)
```

```python
lasso_cv.fit(x_train,y_train)
```

```python
y_pred2 = lasso_cv.predict(x_test)
```

```python
y_pred2
```

```python
mean_squared_error(y_test,y_pred2)
```

```python
mean_absolute_error(y_test,y_pred2)
```

```python
r2_score(y_test,y_pred2)
```

```python
lasso_cv.best_estimator_
```

```python
lasso3 = Lasso(alpha=0.001)
```

```python
lasso3.fit(x_train,y_train)
```

```python
lasso3.intercept_
```

```python
coef = lasso3.coef_
```
