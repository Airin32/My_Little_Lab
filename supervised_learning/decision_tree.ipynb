# Decision Tree


```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
 
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data=data['data'],columns=data['feature_names'])
```

```python
df
```

```python
x = df.copy()
y = data['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=21)
```

```python
clf = DecisionTreeRegressor()
```

```python
clf.get_params()
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
y_predict
```

```python
accuracy_score(y_test,y_predict)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```

```python
clc = DecisionTreeClassifier()
clc = clc.fit(x_train,y_train)
```

```python
featurs_name = x.columns
featurs_name
```

```python
clc.feature_importances_
```

```python
featurs_importance = pd.DataFrame(clc.feature_importances_,index=featurs_name)
featurs_importance
```

```python
featurs_importance.head(10).plot(kind='bar')
```

```python
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clc,
                   feature_names= featurs_name,
                   class_names={0 :'Malignant',1 : 'Benign'},
                   filled=True,
                   fontsize=12)
```
