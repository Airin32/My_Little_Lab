```python
from sklearn.linear_model import LinearRegression
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
```

```python
x1 = np.array([np.random.randint(1,12) for i in range(100)])
y = np.array([np.random.randint(3500,8500) for i in range(100)])
```

```python
x1
```

```python
model_x1  = LinearRegression(n_jobs=-2)
```

```python
data = {
    'Data' : y,
    'Range_time' : x1
}
```

```python
df_x1 = pd.DataFrame(data=data)
```

```python
df_x1
```

```python
x1_tt = df_x1.iloc[:,1].values.reshape(-1,1)
y_tt = df_x1.iloc[:,0].values.reshape(-1,1)
```

```python
x1_train,x1_test,y_train,y_test = train_test_split(x1_tt,y_tt,test_size=0.3,random_state=12)
```

```python
model_x1.fit(x1_tt,y_tt)
```

```python
y_pred = model_x1.predict(x1_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
plt.scatter(x1,y)
plt.plot(x1_test,y_pred,'r')
```
