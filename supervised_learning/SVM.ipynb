```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
mean1 = 55 
std_devl = 10 
num_samples = 500

collum1_numbers = np.random.normal(mean1,std_devl,num_samples)
collum1_numbers = np.clip(collum1_numbers,30,120)
collum1_numbers = np.round(collum1_numbers).astype(int)
mean2 = 18 
std_dev2 = 3 
collumn2_numbers = np.random.normal(mean2,std_dev2,num_samples)
collumn2_numbers = np.clip(collumn2_numbers,12,26)
collumn2_numbers = np.round(collumn2_numbers).astype(int)

collumn3_numbers = np.random.randint(2,size=num_samples)
collumn3_numbers[collum1_numbers > mean1] = 1
data = {
    'miles_per_week' : collum1_numbers,
    'Farthest_run' : collumn2_numbers,
    'Qualited_boston_marathon' : collumn3_numbers
}
df = pd.DataFrame(data=data)
```

```python
df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['miles_per_week'],df['Farthest_run'],c=df['Qualited_boston_marathon'],cmap='viridis')
plt.xlabel('miles_per_week')
plt.ylabel('Farthest_run')
plt.title('Scatter plot')
plt.colorbar(label='Qualified Boston Marathon')
plt.show()
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,2]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=33,test_size=0.2)
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model_reg0 = SVC(C=0.1)
```

```python
model_reg0.fit(x_train,y_train)
```

```python
model_reg0.score(x_test,y_test)
```

```python
model_reg1 = SVC(C=1000)
model_reg1.fit(x_train,y_train)
model_reg1.score(x_test,y_test)
```

```python
model_gamma0 = SVC(gamma=1,C=1)
model_gamma0.fit(x_train,y_train)
model_gamma0.score(x_test,y_test)
```
