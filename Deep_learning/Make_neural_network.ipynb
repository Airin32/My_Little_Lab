```python
import tensorflow as tf
from tensorflow import keras
import os 
os.environ['TFF_CPP_MIN_LOG_LEVEL'] = '2'
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
x_train = x_train.reshape(-1,28*28).astype("float32") / 255.0 
x_test = x_test.reshape(-1,28*28).astype("float32") / 255.0
```

```python
model = keras.Sequential([
    keras.layers.Dense(512,activation='relu'),
    keras.layers.Dense(256,activation='relu'),
    keras.layers.Dense(10)
])
model.compile(
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)
```

```python
model.fit(x_train,y_train,batch_size=32,verbose=2,epochs=5)
```

```python
model.evaluate(x_test,y_test,batch_size=32,verbose=2)
```

```python
model3 = keras.Sequential()
model3.add(keras.Input(shape=(784,)))
model3.add(keras.layers.Dense(512,activation='relu'))
model3.add(keras.layers.Dense(256,activation='relu'))
model3.add(keras.layers.Dense(10))
```

```python
inputs = keras.Input(shape=(784,))
x = keras.layers.Dense(512,activation='relu',name='first_layers')(inputs)
x = keras.layers.Dense(256,activation='relu',name='second_layers')(x)
outputs = keras.layers.Dense(10,activation='softmax')(x)
```

```python
model4 = keras.Model(inputs=inputs,outputs=outputs)
```

```python
model4.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics = ['accuracy']
)
```

```python
model4.fit(x_train,y_train,batch_size=32,epochs=5,verbose=2)
```

```python
feature= model4.predict(x_test)
```
