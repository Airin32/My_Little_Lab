```python
import numpy as np
import pandas as pd
import random

# Array aktivitas
activities = np.array([
    'Drive Gpu', 'Drive OS', 'Drive Ram', 'Drive Disk', 'Drive Monitor', 
    'Music player', 'Video player', 'Chrome', 'Fire fox', 'File Manager', 
    'Microsoft Software', 'Defender', 'Fire wall', 'Setting', 'Left 4 dead', 
    'Video Editor', 'IDE Programming', 'Microsoft Office', 'Foto Editor', 
    'GNS3', 'Virtual Machine', 'Cisco', 'Game'
])

# Waktu aktivitas (Pagi, Siang, Malam)
time_periods = ['Pagi', 'Siang', 'Malam']

# Fungsi untuk membuat dataset pola kebiasaan pengguna
def generate_user_habits_multiactivity(num_samples=1000):
    dataset = []
    for _ in range(num_samples):
        # Pilih waktu (pagi, siang, malam)
        time_period = random.choice(time_periods)
        # Pilih beberapa aktivitas untuk waktu tersebut
        num_activities = random.randint(3, 10)  # Setiap waktu punya 3-10 aktivitas
        selected_activities = random.sample(list(activities), num_activities)
        for activity in selected_activities:
            dataset.append([activity, time_period])
    return pd.DataFrame(dataset, columns=["Activity", "Time_Period"])

# Membuat dataset simulasi
num_samples = 5000  # Sesuaikan jumlah data yang ingin dibuat
dataset = generate_user_habits_multiactivity(num_samples)

# Simpan dataset ke file CSV
dataset.to_csv("user_habits_multiactivity_dataset.csv", index=False)
print("Dataset berhasil dibuat dan disimpan ke 'user_habits_multiactivity_dataset.csv'.")
```

```python
protocol_trust = np.array(['Tcp' for i in range(500)])
Local_Ip_trust = ['192.168.56.1','192.168.7.101','.127.0.0.1']
LocalIp_Trust = np.array([np.random.choice(Local_Ip_trust) for i in range(500)])
```

```python
Protocol_dont_Trust = ['Ftp','Telnet','NBT','SMB','HTTPS','BitTorrent','C&C']
dont_trust_ip = [1000,2000,3000]
dont_trust_ip_foreign = [10000,20000,30000]
```

```python
Security_level_high= np.array([np.random.randint(8,10) for i in range(500)])
iterator_warning_high = np.array([np.random.randint(1,10) for i in range(500)])
security_label = []
for i in range(500) :
    if iterator_warning_high[i] > 7:
        security_label.append(6)
    elif iterator_warning_high[i] >3 and iterator_warning_high[i]<= 7:
        security_label.append(5)
    else :
        security_label.append(4)
```

```python
Security_level_low= np.array([np.random.randint(1,3) for i in range(500)])
iterator_warning_low = np.array([np.random.randint(1,10) for i in range(500)])
for i in range(500) :
    if iterator_warning_low[i] > 5:
        security_label.append(3)
    elif iterator_warning_low[i] >3 and iterator_warning_low[i]<= 5:
        security_label.append(2)
    else :
        security_label.append(1)
```
