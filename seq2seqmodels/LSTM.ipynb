```python
import tensorflow as tf 
import keras 
from sklearn.model_selection import train_test_split
import numpy as np 
```

```python
def random () :
    return [np.random.randint(10) for i in range(10)]
```

```python
X = [random() for _ in range(2000)]
Y =[]
counter = 9
temp = []
for i in range(len(X)):
    for j in range(len(X[i])):
        temp.append(X[i][counter])
        counter-=1
    counter = 9 
    Y.append(temp)
    temp=[]
```

```python
XS = keras.utils.to_categorical(X,num_classes=10)
YS = keras.utils.to_categorical(Y,num_classes=10)
```

```python
x_train,x_test,y_train,y_test = train_test_split(XS,YS,train_size=0.75)
```

```python
lstm_models = keras.Sequential([
    keras.Input(shape=(10,10)),
    keras.layers.LSTM(32),
    keras.layers.RepeatVector(10),
    keras.layers.Dense(10,activation='softmax')
])
```

```python
lstm_models.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
lstm_models.fit(XS,YS,verbose=1,epochs=100)
```

```python
Multi_LSTM = keras.Sequential([
    keras.Input(shape=(10,10)),
    keras.layers.LSTM(64),
    keras.layers.RepeatVector(10),
    keras.layers.LSTM(64),
    keras.layers.RepeatVector(10),
    keras.layers.Dense(10,activation='softmax')
])
```

```python
Multi_LSTM.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
Multi_LSTM.fit(x_train,y_train,verbose=2,epochs=100)
```

```python
LSTM_re_sequences = keras.Sequential([
    keras.Input(shape=(10,10)),
    keras.layers.LSTM(64,return_sequences=True),
    keras.layers.LSTM(32,return_sequences=True),
    keras.layers.TimeDistributed(keras.layers.Dense(10,activation='softmax'))
])
```

```python
LSTM_re_sequences.compile(loss= "categorical_crossentropy",optimizer = "adam",metrics=["accuracy"])
```

```python
LSTM_re_sequences.fit(x_train,y_train,verbose=2,epochs=100)
```

```python
data = LSTM_re_sequences.predict(x_test)
```

```python
Input = keras.Input(shape=(10,10))
lstm1 = keras.layers.LSTM(64,return_state=True)
lstm1_output,state_c,state_h = lstm1(Input)
states = [state_c,state_h]
repeat = keras.layers.RepeatVector(10)
LSTM_output = repeat(lstm1_output)
lstm2 = keras.layers.LSTM(64,return_sequences=True)
all_state_h = lstm2(LSTM_output,initial_state=states)
dense = keras.layers.TimeDistributed(keras.layers.Dense(10,activation='softmax'))
output = dense(all_state_h)
```

```python
Model_LSTM_perfect = keras.Model(Input,output)
```

```python
Model_LSTM_perfect.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
Model_LSTM_perfect.fit(x_train,y_train,verbose=2,epochs=100)
```

```python
data = Model_LSTM_perfect.predict(x_test)
```
