```python
import tensorflow as tf
import keras
import numpy as np 
```

```python
def make_random() :
    return [np.random.randint(0,9) for i in range(4)]
```

```python
x = [make_random() for i in range(1000)]
```

```python
y = []
temp = []
counter = 3
for i in range(len(x)):
    for j in range(len(x[i])):
        temp.append(x[i][counter])
        counter -=1
    counter =3
    y.append(temp)
    temp = []
```

```python
x_train = keras.utils.to_categorical(x,num_classes=10)
y_train = keras.utils.to_categorical(y,num_classes=10)
```

# Make decoder encoder models With Teacher Force 

```python
encoder_inputs = keras.Input(shape=(4,10))
encoder_lstm = keras.layers.LSTM(16,return_state=True)
lstm_ouput,state_h,state_c = encoder_lstm(encoder_inputs)
states = [state_h,state_c]

decoder_input = keras.Input(shape=(None,10))
decoder_lstm = keras.layers.LSTM(16,return_sequences=True,return_state=True)
decoder_output,_,_ = decoder_lstm(decoder_input,initial_state=states)

decoder_dense = keras.layers.Dense(10,activation='softmax')
decoder_output = decoder_dense(decoder_output)

encoder_decoder_models = keras.Model([encoder_inputs,decoder_input],decoder_output)
```

```python
encoder_decoder_models.compile(loss='categorical_crossentropy',optimizer=keras.optimizers.Adam(learning_rate=0.01),metrics=["accuracy"])
```

```python
decoder_state_h_input = keras.Input(shape=(16,))
decoder_state_c_input = keras.Input(shape=(16,))
decoder_states_input = [decoder_state_h_input,decoder_state_h_input]

decoder_outputs,states_h,state_c = decoder_lstm(decoder_input,initial_state=decoder_states_input)
decoder_states = [states_h,state_c]
decoder_outputs = decoder_dense(decoder_outputs)

```

```python
decoder_models = keras.Model(inputs=[decoder_input] + decoder_states_input,outputs = [decoder_outputs] + decoder_states)
```

```python
decoder_models.summary()
```

```python
def decode_sequence (seq_input) :
    states_values = encoder_models.predict(seq_input)
    target_seq = np.zeros((1,1,10))
    target_seq[0,0,0] = 1
    stop_condition = False 
    decoded_seq = list()
    data_sample = list()
    while not stop_condition :
        output_token,h,c = decoder_models.predict([target_seq] + states_values)
        sample_token_index = np.argmax(output_token[0,-1,:])
        data_sample.append(output_token)
        sample_digit = sample_token_index
        decoded_seq.append(sample_digit)
        if (sample_digit == '\n' or len(decoded_seq) == 4):
            stop_condition = True
        target_seq = np.zeros((1,1,10))
        target_seq [0,0,sample_token_index] = 1
        states_values = [h,c]
    return decoded_seq,data_sample
```
